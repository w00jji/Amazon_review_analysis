```python
import pandas as pd
import tqdm
from bs4 import BeautifulSoup
import time
from selenium import webdriver

# Selenium ì›¹ ë“œë¼ì´ë²„ ì´ˆê¸°í™” (Chromeì˜ ê²½ìš°)
driver = webdriver.Chrome()

# ë¦¬ë·° ìˆ˜ì§‘ì„ ìœ„í•œ ë³€ìˆ˜ ì´ˆê¸°í™”
reviews = []
stars = ['five', 'four', 'three', 'two', 'one']  # ë³„ì  ë¦¬ìŠ¤íŠ¸
max_reviews = 101  # ê° ë³„ì ë‹¹ ìµœëŒ€ ë¦¬ë·° ê°œìˆ˜ ì„¤ì •
max_pages = 10  # ìµœëŒ€ í˜ì´ì§€ ìˆ˜ ì„¤ì •

# ë³„ì  ë§¤í•‘ì„ ìœ„í•œ ë”•ì…”ë„ˆë¦¬
star_mapping = {'five': 5, 'four': 4, 'three': 3, 'two': 2, 'one': 1}

# ë³„ì  ë³„ë¡œ ë¦¬ë·° ìˆ˜ì§‘
for star in tqdm.tqdm(stars, desc="Collecting reviews"):  # tqdmì„ ì‚¬ìš©í•˜ì—¬ ì§„í–‰ ìƒí™© í‘œì‹œ
    page_number = 1  # í˜ì´ì§€ ë²ˆí˜¸ ì´ˆê¸°í™”
    star_reviews = []  # í•´ë‹¹ ë³„ì ì˜ ë¦¬ë·°ë¥¼ ì €ì¥í•  ë¦¬ìŠ¤íŠ¸

    while len(star_reviews) < max_reviews and page_number <= max_pages:  # ìµœëŒ€ í˜ì´ì§€ ìˆ˜ì™€ ë¦¬ë·° ìˆ˜ ì œí•œ
        # Amazon ë¦¬ë·° í˜ì´ì§€ URL
        url = f'https://www.amazon.com/dalba-Piedmont-Light-weight-Moisturizer-Comedogenic/product-reviews/B0BFQ9RD5B/ref=cm_cr_arp_d_viewopt_sr?ie=UTF8&reviewerType=all_reviews&pageNumber={page_number}&filterByStar={star}_star'
        driver.get(url)
        time.sleep(3)  # í˜ì´ì§€ ë¡œë”© ëŒ€ê¸°

        # ë¦¬ë·° ìš”ì†Œ ì°¾ê¸°
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        review_elements = soup.find_all("span", {'data-hook': 'review-body'})

        # ë¦¬ë·° í…ìŠ¤íŠ¸ ì¶”ì¶œ
        for review in review_elements:
            # starë¥¼ ìˆ«ìë¡œ ë³€í™˜í•˜ì—¬ ì¶”ê°€
            star_reviews.append({'Review': review.get_text(strip=True), 'Star': star_mapping[star]})  # ë¦¬ë·°ì™€ ë³„ì  ì¶”ê°€
            # print(f"Star: {star_mapping[star]}, Review: {review.get_text(strip=True)}")  # ë³„ì ê³¼ ë¦¬ë·° ì¶œë ¥
            if len(star_reviews) >= max_reviews:  # ìµœëŒ€ ë¦¬ë·° ìˆ˜ë¥¼ ì´ˆê³¼í•˜ë©´ ì¢…ë£Œ
                break

        # ë‹¤ìŒ í˜ì´ì§€ë¡œ ì´ë™
        page_number += 1

    # í•´ë‹¹ ë³„ì ì˜ ë¦¬ë·°ë¥¼ ì „ì²´ ë¦¬ë·° ë¦¬ìŠ¤íŠ¸ì— ì¶”ê°€
    reviews.extend(star_reviews)

# ë¦¬ë·° ë°ì´í„°í”„ë ˆì„ìœ¼ë¡œ ë³€í™˜
df = pd.DataFrame(reviews)

# ë°ì´í„°í”„ë ˆì„ì„ TSV íŒŒì¼ë¡œ ì €ì¥
df.to_csv('amazon_reviews.tsv', sep='\t', index=False)

# ë“œë¼ì´ë²„ ì¢…ë£Œ
driver.quit()

df
```

```python
import pandas as pd

# íŒŒì¼ ê²½ë¡œ
files = [
    'amazon_reviews_1star.tsv',
    'amazon_reviews_2star.tsv',
    'amazon_reviews_3star.tsv',
    'amazon_reviews_4star.tsv',
    'amazon_reviews_5star.tsv'
]

# ë¹ˆ ë¦¬ìŠ¤íŠ¸ì— ë°ì´í„°í”„ë ˆì„ì„ ì €ì¥í•  ê²ƒ
dfs = []

# íŒŒì¼ë³„ë¡œ ë°ì´í„° ì½ê³ , ë³„ì  ì»¬ëŸ¼ ì¶”ê°€
for i, file in enumerate(files, 1):
    # ë°ì´í„° ì½ê¸°
    df = pd.read_csv(file, sep='\t')
    
    # ë³„ì  ì»¬ëŸ¼ ì¶”ê°€
    df['Star'] = i  # ë³„ì ì— ë§ëŠ” 1, 2, 3, 4, 5 ì¶”ê°€
    
    # ë°ì´í„°í”„ë ˆì„ ë¦¬ìŠ¤íŠ¸ì— ì¶”ê°€
    dfs.append(df)

# ëª¨ë“  ë°ì´í„°í”„ë ˆì„ ë³‘í•©
final_df = pd.concat(dfs, ignore_index=True)

# ë³‘í•©ëœ ë°ì´í„° í™•ì¸
print(final_df)

# ë³‘í•©ëœ ë°ì´í„°ë¥¼ TSV íŒŒì¼ë¡œ ì €ì¥
final_df.to_csv('amazon_reviews_all.tsv', sep='\t', index=False)
```
