```python
import pandas as pd
import tqdm
from bs4 import BeautifulSoup
import time
from selenium import webdriver

# Selenium 웹 드라이버 초기화 (Chrome의 경우)
driver = webdriver.Chrome()

# 리뷰 수집을 위한 변수 초기화
reviews = []
stars = ['five', 'four', 'three', 'two', 'one']  # 별점 리스트
max_reviews = 101  # 각 별점당 최대 리뷰 개수 설정
max_pages = 10  # 최대 페이지 수 설정

# 별점 매핑을 위한 딕셔너리
star_mapping = {'five': 5, 'four': 4, 'three': 3, 'two': 2, 'one': 1}

# 별점 별로 리뷰 수집
for star in tqdm.tqdm(stars, desc="Collecting reviews"):  # tqdm을 사용하여 진행 상황 표시
    page_number = 1  # 페이지 번호 초기화
    star_reviews = []  # 해당 별점의 리뷰를 저장할 리스트

    while len(star_reviews) < max_reviews and page_number <= max_pages:  # 최대 페이지 수와 리뷰 수 제한
        # Amazon 리뷰 페이지 URL
        url = f'https://www.amazon.com/dalba-Piedmont-Light-weight-Moisturizer-Comedogenic/product-reviews/B0BFQ9RD5B/ref=cm_cr_arp_d_viewopt_sr?ie=UTF8&reviewerType=all_reviews&pageNumber={page_number}&filterByStar={star}_star'
        driver.get(url)
        time.sleep(3)  # 페이지 로딩 대기

        # 리뷰 요소 찾기
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        review_elements = soup.find_all("span", {'data-hook': 'review-body'})

        # 리뷰 텍스트 추출
        for review in review_elements:
            # star를 숫자로 변환하여 추가
            star_reviews.append({'Review': review.get_text(strip=True), 'Star': star_mapping[star]})  # 리뷰와 별점 추가
            # print(f"Star: {star_mapping[star]}, Review: {review.get_text(strip=True)}")  # 별점과 리뷰 출력
            if len(star_reviews) >= max_reviews:  # 최대 리뷰 수를 초과하면 종료
                break

        # 다음 페이지로 이동
        page_number += 1

    # 해당 별점의 리뷰를 전체 리뷰 리스트에 추가
    reviews.extend(star_reviews)

# 리뷰 데이터프레임으로 변환
df = pd.DataFrame(reviews)

# 데이터프레임을 TSV 파일로 저장
df.to_csv('amazon_reviews.tsv', sep='\t', index=False)

# 드라이버 종료
driver.quit()

df
```

#### 평점 합치기

```python
import pandas as pd

# 파일 경로
files = [
    'amazon_reviews_1star.tsv',
    'amazon_reviews_2star.tsv',
    'amazon_reviews_3star.tsv',
    'amazon_reviews_4star.tsv',
    'amazon_reviews_5star.tsv'
]

# 빈 리스트에 데이터프레임을 저장할 것
dfs = []

# 파일별로 데이터 읽고, 별점 컬럼 추가
for i, file in enumerate(files, 1):
    # 데이터 읽기
    df = pd.read_csv(file, sep='\t')
    
    # 별점 컬럼 추가
    df['Star'] = i  # 별점에 맞는 1, 2, 3, 4, 5 추가
    
    # 데이터프레임 리스트에 추가
    dfs.append(df)

# 모든 데이터프레임 병합
final_df = pd.concat(dfs, ignore_index=True)

# 병합된 데이터 확인
print(final_df)

# 병합된 데이터를 TSV 파일로 저장
final_df.to_csv('amazon_reviews_all.tsv', sep='\t', index=False)
```
